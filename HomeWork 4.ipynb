```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def circle(x, y, r):
    x1 = []
    y1 = []
    for i in range(360):
        theta = 2 * np.pi * i / 360
        x1.append(x + r*np.cos(theta)) #리스트에 차례대로 원소를 삽입
        y1.append(y + r*np.sin(theta)) #append문 사용
    return x1, y1
xx, yy = circle(5, 10, 5)
plt.figure(figsize = (5, 5))
plt.plot(xx, yy)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-5.0, 15.0)
plt.ylim(0, 20)
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def f1(x, tau):
    y = np.zeros([len(x)])
    for i in range(len(x)):
        value = np.round(x[i])
        dist = np.abs(x[i] - value)
        if dist <= tau/2:
            y[i] = 1
        else:
            y[i] = 0
    return y

x = np.linspace(0, 5, 1000)
tau = 0.3
y = f1(x, tau)
plt.plot(x, y)

plt.grid(True)
plt.show()
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def f1(x, tau):
    y = np.zeros([len(x)])
    for i in range(len(x)):
        value = np.round(x[i])
        dist = np.abs(x[i] - value)
        if dist <= tau/2:
            y[i] = 1
        else:
            y[i] = 0
    return y

def f2(x, tau, f):
    yt = np.zeros([len(x)])
    for i in range(len(x)):
        yt[i] = np.cos(2 * np.pi * f * x[i] + np.pi * f1(x, tau)[i])
    return yt
x = np.linspace(0, 2, 1000)
tau = 0.5
f = 5
y1 = f1(x, tau)
y2 = f2(x, tau, f)

plt.subplot(2, 1, 1)
plt.plot(x, y1)
plt.grid(True)
plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.grid(True)
plt.show()
```
