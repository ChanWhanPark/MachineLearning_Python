```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def circle(x, y, r):
    x1 = []
    y1 = []
    for i in range(360):
        theta = 2 * np.pi * i / 360
        x1.append(x + r*np.cos(theta)) #리스트에 차례대로 원소를 삽입
        y1.append(y + r*np.sin(theta)) #append문 사용
    return x1, y1
xx, yy = circle(5, 10, 5)
plt.figure(figsize = (5, 5))
plt.plot(xx, yy)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-5.0, 15.0)
plt.ylim(0, 20)
plt.grid(True)
plt.show()
```
